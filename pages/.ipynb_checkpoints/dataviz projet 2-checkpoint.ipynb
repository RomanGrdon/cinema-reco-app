```python
#import data
df = pd.read_csv('/content/drive/MyDrive/projet 2/data_P2_final.csv')
df.head(3)
```

```python
df.info()
```

```python
# s'assurer que les colonnes sont bien numériques
df["release_year"] = pd.to_numeric(df["release_year"], errors="coerce")
df["runtimeMinutes"] = pd.to_numeric(df["runtimeMinutes"], errors="coerce")

# suppression des lignes invalides
df_clean = df.dropna(subset=["release_year", "runtimeMinutes"])
```

```python
df_clean["year_5"] = (df_clean["release_year"] // 5) * 5
```

```python
df_clean = df_clean[df_clean["release_year"] >= 1960]
```

```python
runtime_5years = (
    df_clean
    .groupby("year_5")["runtimeMinutes"]
    .mean()
    .reset_index()
    .sort_values("year_5")
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.bar(
    runtime_5years["year_5"],
    runtime_5years["runtimeMinutes"],
    width=4  # largeur adaptée à une période de 5 ans
)

plt.xlabel("année")
plt.ylabel("Durée moyenne (minutes)")
plt.title("Durée moyenne des films")

plt.xticks(runtime_5years["year_5"], rotation=45)

plt.show()
```

```python
actor_counts = (
    df
    .groupby("primaryName")
    .size()
    .reset_index(name="nb_films")
    .sort_values("nb_films", ascending=False)
)
```

```python
top_actors = actor_counts.head(20)
top_actors
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.barh(
    top_actors["primaryName"],
    top_actors["nb_films"]
)

plt.xlabel("Nombre de films")
plt.title("Top 20 des acteurs ")

plt.gca().invert_yaxis()  #ascending

plt.show()
```

```python
df_ratings = df[df["vote_count"] >= 500]
```

```python
top_movies = (
    df_ratings
    .sort_values("vote_average", ascending=False)
    .head(20)
)
top_movies
```

```python
plt.figure(figsize=(10, 6))

bars = plt.barh(
    top_movies["title"],
    top_movies["vote_average"],
    color="skyblue"
)

plt.xlabel("Note moyenne")
plt.ylabel("Film")
plt.title("Top 20 des films les mieux notés (≥ 500 votes)")

plt.gca().invert_yaxis()  # film le mieux noté en haut

# Ajouter la note sur chaque barre
for bar, note in zip(bars, top_movies["vote_average"]):
    plt.text(
        bar.get_width() + 0.05,  # un peu à droite de la barre
        bar.get_y() + bar.get_height()/2,
        f"{note:.1f}",  # une décimale
        va='center'
    )

plt.show()
```
